```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig2, ax2 = plt.subplots()

x_values = list(range(1,5+1))
y_values = [x**3 for x in x_values]

ax.scatter(x_values, y_values, linewidth=3)

x_values2 = list(range(1,5000+1))
y_values2 = [x**3 for x in x_values2]

ax2.scatter(x_values2, y_values2, linewidth=3)

```
